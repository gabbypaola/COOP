#2017 Data

```python
race_2017 = gc.open('RACE2017').sheet1

# get_all_values gives a list of rows.
rows9 = race_2017.get_all_values()
#print(rows)

race_data = pd.DataFrame.from_records(rows9)


new_header = race_data.iloc[1] #grab the first row for the header
race_data = race_data[2:] #take the data less the header row
race_data.columns = new_header #set the header row as the df header

filter = ['Label','Percent']

race_data = race_data[race_data.columns[race_data.columns.isin(filter)]]

new_header = race_data.iloc[0] #grab the first row for the header
race_data = race_data[1:] #take the data less the header row
race_data.columns = new_header #set the header row as the df header
```

```python
column_names = []
for i in race_data.columns[1:]:
  column_names.append(i[6:11])

original_col_names = list(race_data.columns[1:])

col_names_dict = {original_col_names[i]: column_names[i] for i in range(len(original_col_names))}

race_data= race_data.rename(columns=col_names_dict)

race17 = race_data.loc[[3,4,38,39,41,46]]

race17 = race17.melt(id_vars=["Label (Grouping)"], 
        var_name="Zip Code", 
        value_name="Percent")

race17 = race17.rename(columns = {'Label (Grouping)':'Demographic'})
```

```python
race17['Year'] = 2017
```

```python
race17['Year']=race17['Year'].astype(str).astype(int)
```

```python
# 19,310,616 rows of data
race17
```

```python
race17['Zip Code']=race17['Zip Code'].astype(str).astype(int)
```
